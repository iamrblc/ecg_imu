```python
def to_camel_case(text):
    import re
    if text == "":
        result = ""
    else:
        splittext = re.split(r'[_-]', text)
        changed = []
        changed.append(splittext[0].lower())
        for i in range(1,len(splittext)):
            changed.append(splittext[i].capitalize())
        result = "".join(changed)
    return result

def to_camel_case(text):
    return text[:1] + text.title()[1:].replace('_', '').replace('-', '')
```
